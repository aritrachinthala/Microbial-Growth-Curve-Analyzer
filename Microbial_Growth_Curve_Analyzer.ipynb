```python
# Microbial Growth Curve Analyzer
# Plots microbial growth curve and calculates growth rate

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# INPUT DATA FROM USER
# -----------------------------
# Enter values separated by commas
time = np.array(list(map(float, input(
    "Enter time values (hours/minutes) separated by commas: "
).split(","))))

od = np.array(list(map(float, input(
    "Enter OD600 values separated by commas: "
).split(","))))

# -----------------------------
# LOG PHASE SELECTION
# -----------------------------
log_time = np.array(list(map(float, input(
    "Enter log phase time values separated by commas: "
).split(","))))

log_od = np.array(list(map(float, input(
    "Enter log phase OD600 values separated by commas: "
).split(","))))

# -----------------------------
# CALCULATIONS
# -----------------------------
# Convert OD to natural log
ln_od = np.log(log_od)

# Calculate growth rate (μ)
slope, intercept = np.polyfit(log_time, ln_od, 1)
growth_rate = slope

# Doubling time
doubling_time = np.log(2) / growth_rate

# -----------------------------
# PLOTTING
# -----------------------------
plt.figure()
plt.plot(time, od, marker='o', label='Growth Curve')
plt.plot(log_time, log_od, marker='o', label='Log Phase')
plt.xlabel("Time (hours)")
plt.ylabel("Optical Density (OD600)")
plt.title("Microbial Growth Curve")
plt.legend()
plt.grid()
plt.show()

# -----------------------------
# OUTPUT RESULTS
# -----------------------------
print("\n----- Microbial Growth Analysis -----")
print(f"Specific Growth Rate (μ): {growth_rate:.3f} per hour")
print(f"Doubling Time: {doubling_time:.2f} hours")

# values of actual exp done by me  to test run the code
# time(minutes):0, 20, 40, 60, 80, 100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300, 320, 340, 360, 380, 400, 420, 440, 460, 480, 500, 520, 540
# OD values :0.00, 0.09, 0.11, 0.13, 0.14, 0.16, 0.17, 0.18, 0.21, 0.22, 0.30, 0.42, 0.46, 0.49, 0.50, 0.53, 0.56, 0.60, 0.63, 0.68, 0.70, 0.74, 0.77, 0.80, 0.80, 0.80, 0.52, 0.46
```
